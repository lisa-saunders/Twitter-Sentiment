# Tweets w/o Geo Location

```python
#Importing other libraries
import pandas as pd
import numpy as np
import snscrape.modules.twitter as sntwitter
import itertools
import time
```

```python
df = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    '(paris agreement) OR (climate change) OR (global warming) OR (climate action) OR (climate hoax) OR (climate trump) OR (climate deniers) OR (greta thunberg climate) lang:en').get_items(), 5))[['id', 'date', 'content', 'coordinates', 'place', 'user']]
```

```python
count = 0

for i in range(1, 20000):
    count += 1
    if count < 800:
        print('Starting iterations')
        mx = df['id'].iloc[-1]
        df1 = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
            '(paris agreement) OR (climate change) OR (global warming) OR (climate action) OR (climate hoax) OR (climate trump) OR (climate deniers) lang:en max_id:%s' % mx).get_items(), 2))[['id', 'date', 'content', 'coordinates', 'place', 'user']]
        df1 = df1.iloc[1:]
        df = df.append(df1, ignore_index=True)
        print('Iteration #', i, 'done.')
        time.sleep(1)
    else:
        print('going to sleep')
        time.sleep(600)
        count = 0
        print('waking up')
        continue
```
