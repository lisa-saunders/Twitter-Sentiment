```python
#Importing other libraries
import pandas as pd
import numpy as np
import snscrape.modules.twitter as sntwitter
import itertools
```

```python
timeout = time.time() + 60*5
```

```python
#hashtags: #climate OR #climatechange OR #globalwarming OR #environment OR #climateaction OR #climatehoax OR #parisagreement
```

```python
df = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
    '#climatechange place:96683cc9126741d1').get_items(), 50))[['id', 'date', 'content', 'coordinates', 'place', 'user']]

for i in range(1, 20):
    print('Starting iterations')
    mx = df['id'].iloc[-1]
    df1 = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper(
        '#climatechange place:96683cc9126741d1 max_id:%s' % mx).get_items(), 10))[['id', 'date', 'content', 'coordinates', 'place', 'user']]
    df1 = df1.iloc[1:]
    df = df.append(df1, ignore_index=True)
    print('Iteration #', i, 'done.')
    if time.time() > timeout:
        print('Timeout')
        break
```
